```python
'''
利用request下载页面
自动检测页面编码
'''

import urllib
import chardet

if __name__ == '__main__':
    url = 'https://blog.csdn.net/qq_42247018/article/details/88762145'
    rsp = urllib.request.urlopen(url)
    html = rsp.read()
    #利用 chardet自动检测
    code_type = chardet.detect(html)
    print(type(code_type))
    print(code_type)
    #使用get取值保证不会出错
    html = html.decode(code_type.get("encoding","utf-8"))#先检测，如果没有检测到就用utf-8
    print(html)
```
